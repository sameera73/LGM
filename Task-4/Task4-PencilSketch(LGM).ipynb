# Image to Pencil Sketch with Python

```python
import cv2
from matplotlib import pyplot as plt
```

```python
sam=cv2.imread('C:/Users/Sameera/Documents/LGM/Task4-PencilSketch/krishna.jpeg',0)
```

```python
plt.imshow(sam)
```

```python
sam1=cv2.cvtColor(sam, cv2.COLOR_BGR2RGB) #converts bgr color to rgb color
```

```python
plt.imshow(sam1)
```

```python
inv=255-sam1
```

```python
plt.imshow(inv)
```

```python
blur=cv2.GaussianBlur(inv, (111,111),0)
```

```python
inv1=255-blur
s=cv2.divide(sam1,inv1,scale=255)
```

```python
plt.imshow(s)
```
